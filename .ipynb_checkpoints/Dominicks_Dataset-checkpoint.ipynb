```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

# Предсказание спроса на пиво с использованием Dominick’s dataset

## Анализ данных

```python
sales_df = pd.read_parquet("beer_sales_data.parquet")
display(sales_df.head())
```

```python
upc_df = pd.read_parquet("beer_upc.parquet")
display(upc_df.head())
```

Объединяем таблицы для пива по общему столбцу <b>upc</b>.

```python
sales_upc_df = pd.merge(sales_df, upc_df, on="upc")
display(sales_upc_df.head())
```

Удаляем не влияющие на продажи столбцы <b>com_code</b>, <b>nitem</b>, <b>case</b> и создаём новый столбец <b>units_sold</b> для количества проданных продуктов согласно формуле из [инструкции](https://www.chicagobooth.edu/research/kilts/research-data/-/media/enterprise/centers/kilts/datasets/dominicks-dataset/dominicks-manual-and-codebook_kiltscenter.pdf).

```python
sales_upc_df.drop(columns=["com_code", "nitem", "case"], inplace=True)
sales_upc_df["units_sold"] = sales_upc_df.apply(lambda row: row.move/row.qty, axis=1)
sales_upc_df.drop(columns=["qty", "move"], inplace=True)
display(sales_upc_df.head())
```

### График продаж по магазинам

```python
ax = sns.barplot(data=sales_upc_df, x="store", y="units_sold", estimator=sum)
ax.tick_params(axis="x", labelrotation=90, labelsize=5.0)
ax.set_title("Total sales per store")
ax.set_xlabel("Store")
ax.set_ylabel("Sales")
```

### График продаж по размерам тары

```python
ax = sns.barplot(data=sales_upc_df, x="size", y="units_sold", estimator=sum)
ax.tick_params(axis="x", labelrotation=90)
ax.set_title("Total sales per size")
ax.set_xlabel("Size")
ax.set_ylabel("Sales")
```

Размер <b>6/12 O</b> продавался сильнее всего, однако мне не удалось информацию по тому, как перевести размер в более понятный эквивалент по типу объёма в литрах, поэтому в данной задаче он используется только как категориальный признак. Также как категориальный признак используется <b>sale</b>.

```python
sales_upc_df.drop(columns=["descrip"], inplace=True)
sales_upc_df["size"] = sales_upc_df["size"].astype("category").cat.codes
sales_upc_df["sale"] = sales_upc_df["sale"].astype("category").cat.codes
display(sales_upc_df.head())
```

### График продаж по неделям

```python
weekly_sales = sales_upc_df.groupby("week")["units_sold"].sum().reset_index().sort_values("week")
plt.plot(weekly_sales["week"], weekly_sales["units_sold"], marker="o")
plt.title("Weekly sales trends")
plt.xlabel("Week")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

Как видно из графика, наиболее высокие продажи были в районе 240 недели, а после 310 недели резко просели.

### График продаж в каждом магазине по неделям

```python
weekly_sales_by_store = sales_upc_df.groupby(["week", "store"])["units_sold"].sum().unstack().fillna(0).sort_index()
plt.figure()
for store in weekly_sales_by_store.columns:
    plt.plot(weekly_sales_by_store.index, weekly_sales_by_store[store], marker="o", label=store)
plt.title("Weekly sales trends by store")
plt.xlabel("Week")
plt.ylabel("Sales")
plt.legend(title="Store", bbox_to_anchor=(2.8, 1), loc="upper right", ncols=10)
plt.grid(True)
plt.show()
```

Из графика видно, что наибольшее число продаж пришлось на магазины <b>8</b>, <b>9</b> и <b>32</b>.

### График рассеяния продаж и цены

```python
ax = sns.scatterplot(data=sales_upc_df, x="units_sold", y="price", size="units_sold")
ax.set_xlabel("Sales")
ax.set_ylabel("Price")
```

Из графика видно, что в целом при увеличении цены продажи падают.

### Матрица корреляции

```python
ax = sns.heatmap(sales_upc_df.corr(), annot=True)
ax.set_title("Correlation matrix")
```

Согласно матрице корреляции, наиболее коррелирующими по отношению к <b>units_sold</b> являются признаки <b>price</b>, <b>sale</b> и <b>week</b>.

```python
X = sales_upc_df[["price", "sale", "week"]].values
y = sales_upc_df["units_sold"].values
```

## Прогнозирование спроса

В качестве моделей для прогнозирования спроса на пиво я выбрал регрессор на основе дерева решений и линейную регрессию, так как они являются наиболее простыми.

### Масштабирование признаков

```python
scaler = StandardScaler()
X = scaler.fit_transform(X, y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=73)
```

### Модель дерева решений для регрессии

```python
model = DecisionTreeRegressor(random_state=73)
model.fit(X_train, y_train)
print("R2-Score: "+str(model.score(X_test, y_test))+".")
print("Mean squared error: "+str(mean_squared_error(y_test, model.predict(X_test)))+".")
```

### Линейная регрессия

```python
model = LinearRegression(fit_intercept=False, n_jobs=-1)
model.fit(X_train, y_train)
print("R2-Score: "+str(model.score(X_test, y_test))+".")
print("Mean squared error: "+str(mean_squared_error(y_test, model.predict(X_test)))+".")
```

## Выводы

Согласно проведённому анализу выяснилось, что регрессор на основе дерева решений показывает более высокую точность и низкую среднеквадратичную ошибку как на обучающей, так и на тестовой выборке.
